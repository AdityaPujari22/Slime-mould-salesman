```python


pos = (x[0], y[0])


dist = []
for j in range(mt.factorial(len(x) - 1)):
  places = []
  distance = 0
  min_dist = 1
  for i in range(10):
    places.append(pos)

    next_pos_ind = (np.random.randint(len(x)))
    next_pos = (x[next_pos_ind], y[next_pos_ind])
    if next_pos not in places:
      distance += mt.sqrt( (next_pos[0] - pos[0])**2 + (next_pos[1] - pos[1])**2 )
      pos = next_pos
      dist.append(distance)

  if distance < min_dist:
    min_dist = distance
    shortest_path = (places, min_dist)
    
print(shortest_path)


```

```python
path = shortest_path[0]
for i in range(len(path)-1):
  plt.plot((path[i][0], path[i+1][0]), (path[i][1], path[i+1][1]), 'ro-')


print(shortest_path[1])
```

```python
for j in range(mt.factorial(len(x))):
  places = []
  distance = 0
  max_dist = 0
  pos = (x[0], y[0])
  for i in range(10):
    if pos not in places:
      places.append(pos)

    next_pos_ind = (np.random.randint(len(x)))
    next_pos = (x[next_pos_ind], y[next_pos_ind])
    if next_pos not in places:
      distance += mt.sqrt( (next_pos[0] - pos[0])**2 + (next_pos[1] - pos[1])**2 )
      pos = next_pos
      
  if distance > max_dist:
    max_dist = distance
    longest_path = (places, max_dist)
    
print(longest_path)
```

```python
path = longest_path[0]
for i in range(len(path)-1):
  plt.plot((path[i][0], path[i+1][0]), (path[i][1], path[i+1][1]), 'ro-')

print(longest_path[1])
```
